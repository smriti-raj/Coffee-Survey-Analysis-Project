```python
# Create Age Groups (Better Analysis)
# Fixed age_group function with proper age range checks
def age_group(age):
    if pd.isna(age):
        return "Unknown"
    
    # Convert to numeric if it's not already
    try:
        age_val = float(age)
    except (ValueError, TypeError):
        return "Unknown"
    
    # Check age ranges properly
    if 18 <= age_val <= 24:
        return "Young"
    elif 25 <= age_val <= 34:
        return "Adult"
    elif 35 <= age_val <= 44:
        return "Mid-age"
    else:
        return "Senior"

# Apply the corrected function
# No need to convert to string first since our function handles numeric values
df['age_group'] = df['age'].apply(age_group)

# Display the results
df[['age', 'age_group']].head()
```

```python
####  Advanced Group Comparison

# Creating  sample data 
np.random.seed(42)
employment_types = ['Full-time', 'Part-time', 'Student', 'Unemployed', 'Self-employed', 'Retired']
coffee_levels = ['Light', 'Moderate', 'Heavy']
data = {
    'employment': np.random.choice(employment_types, size=200),
    'coffee_level': np.random.choice(coffee_levels, size=200),
    'cups_numeric': np.random.randint(1, 10, size=200)
}
df = pd.DataFrame(data)

# Calculate mean cups by employment and coffee_level
result = df.groupby(['employment', 'coffee_level'])['cups_numeric'].mean().unstack()

# Display the result
print("Mean cups of coffee by employment status and consumption level:")
print(result)

# Create a visualization of the data
plt.figure(figsize=(12, 6))
result.plot(kind='bar')
plt.title('Average Coffee Consumption by Employment Status and Consumption Level')
plt.xlabel('Employment Status')
plt.ylabel('Average Cups per Day')
plt.xticks(rotation=45)
plt.legend(title='Coffee Level')
plt.tight_layout()
plt.show()
```
